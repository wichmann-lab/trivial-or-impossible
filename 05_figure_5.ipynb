```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# General settings
figure_path = "./figures/main/"

# Seaborn settings
sns.set_palette(sns.color_palette("magma"))
sns.set_style("ticks")
sns.set_style({'font.family': 'Lato'})
sns.set_context("paper", font_scale = 1.8)
```

## Figure 5 (a)

```python
from scripts.analysis_helpers import main_plot

# Set base network and conditions
base_network = "Res18"
folder_name = f"./results/{base_network}/"
conditions = [f"{base_network}_Base_condition",
              f"{base_network}_Plus_1ep",
              f"{base_network}_Plus_10ep",
              f"{base_network}_Different_optimizer",
              f"{base_network}_Different_batchsize",
              f"{base_network}_Different_initialisation",
              f"{base_network}_Different_LR",
              f"{base_network}_CUDA_nondeterministic",
              f"{base_network}_Different_dataorder",
              f"{base_network}_Different_architecture",
              f"{base_network}_Half_data",
              f"{base_network}_Combined_condition"]

# Load heat map arrays
heat_array = np.load(folder_name + f"{base_network}_heat_array.npy", allow_pickle=True)
heat_array_rm = np.load(folder_name + f"{base_network}_heat_array_rm.npy", allow_pickle=True)
```

```python
# Figure settings and clean label
plt.figure(figsize=(24, 12))
cond_clear = [cond.replace(base_network, "").replace("_", " ") for cond in conditions]
sns.set(font_scale=2.0)

# Left subplot
plt.subplot(121)
sns.heatmap(heat_array, annot=True, fmt = ".2f", annot_kws={"fontsize":12}, square=True,
            xticklabels=cond_clear, yticklabels=cond_clear, cmap="Blues", cbar=False,
            vmin=0, vmax=1)

# Right subplot
plt.subplot(122)
ax = sns.heatmap(heat_array_rm, annot=True, fmt = ".2f", annot_kws={"fontsize":12}, square=True,
            xticklabels=cond_clear, yticklabels=False, cmap="Blues", cbar_kws={"shrink": .80},
            vmin=0, vmax=1)

# Save figure
plt.tight_layout()
plt.savefig(figure_path + 'figure_5_a.png')
```

## Figure 5 (b)

```python
from scripts.analysis_helpers import main_plot

# Set base network and conditions
base_network = "Rebuttal"
folder_name = f"./results/{base_network}/"
conditions = os.listdir(folder_name)

# Filter .npy arrays from conditions list
conditions = [f for f in conditions if '.npy' not in f]
print(f"{len(conditions)} models found: {conditions}.")

# Load heat map arrays
heat_array = np.load(folder_name + f"{base_network}_heat_array.npy", allow_pickle=True)
heat_array_rm = np.load(folder_name + f"{base_network}_heat_array_rm.npy", allow_pickle=True)
```
